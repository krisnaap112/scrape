```python
for i in products:
    name = i['name']
    price = i['price']
    #imageLink = i['primary_image']['original']
    discount = i['discountPercentage']
    #rating = i['stats']['rating']
    rating = i['rating']
    category = i['categoryName']
    shop_name = i['shop']['name']
    shop_loc = i['shop']['city']
    count_review = i['countReview']
    rating_avg = i['ratingAverage']

    result.append({
        'Name': name,
        'Price': price,
        #'imageLink': imageLink,
        'Discount': discount,
        'Ratings': rating,
        'Category': category,
        'Shop_name': shop_name,
        'Shop_loc': shop_loc,
        'Count_review' : count_review,
        'Rating_avg' : rating_avg,
    })

df = pd.DataFrame.from_dict(result)
df.to_csv('result.csv', encoding='utf-8')

#df=pd.read_json('result.json')
#df.to_csv('result.csv')
```
